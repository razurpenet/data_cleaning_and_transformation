```python
special_chars = ['!', '@', '#', '$', '%', '^', '&', '*', '(', ')', '_', '+', '-', '=', '{', '}', '[', ']', 
                 '|', '\\', ':', ';', '"', "'", '<', '>', '?', ',', '.', '/', ' ']

logs.isin(special_chars).sum()
```

```python
def clean_dataset(col):
    special_chars = ['!', '@', '#', '$', '%', '^', '&', '*', '(', ')', '_', '+', '-', '=', '{', '}', '[', ']', 
                 '|', '\\', ':', ';', '"', "'", '<', '>', '?', ',', '.', '/', ' ']
    if col == str:
        col = col.strip(special_chars)
        return col
```

```python
logs['status'].apply(clean_dataset)
```

```python
logs['status'] = logs['status'].str.replace('\W', 'unknown', regex=True)  #\W regex non-word check
```

```python
logs.head(20)
```

```python
def extract_state(text):
    result = text.split(';')[0]
    return result

def extract_country(text):
    result = text.split(';')[1]
    return result
```

```python
logs['server_state'] = logs['serverregion'].apply(extract_state)
logs['server_country'] = logs['serverregion'].apply(extract_country)
```

```python
logs.head(10)
```

```python
logs = logs.drop(columns=['serverregion'])
```

```python
logs['timestamp'] = pd.to_datetime(logs['timestamp'], unit='s')
```
